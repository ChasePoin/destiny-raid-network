```python
import networkx as nx
from networkx.algorithms.community.quality import modularity
import ndlib.models.ModelConfig as mc
import ndlib.models.epidemics as ep

g = nx.read_gml("nooticing_same_clans2.gml")

# Model selection
model = ep.SIRModel(g)

# Model Configuration
cfg = mc.Configuration()
cfg.add_model_parameter('beta', 0.05) # infection rate
# cfg.add_model_parameter('lambda', 0.005) # recov prob
cfg.add_model_parameter('gamma', 0.005) # removal prob
cfg.add_model_parameter("fraction_infected", 0.0000002) # start with how much infected? 1 person
model.set_initial_status(cfg)

# Simulation execution
iterations = model.iteration_bunch(50)

for iteration in iterations:
    for node in g.nodes(): 
        if node in iteration['status']:
            g.nodes()[node]['infect_status'] = iteration['status'][node]
    print(iteration)

nx.write_gml(g, "infect_iter49_clans_SIR.gml")

```
